```python
%%time
f = open(FAS,'rt')
f.readline()
line = f.readline()
f1 = line.split(';')[0]
print(f1)
out_file = os.path.join(OFAS,f1+'.csv')
of = open(out_file,'wt')
n0, n1, n2 = 0,0,0
while line != '':
    n1+=1
    line = f.readline()
    #print(n1,end='\r')
    if len(line) >1:       
        f2 = line.split(';')[0]
        f2 = str(f2).replace("/", " ")
        f2 = str(f2).replace("\\", " ")        
        if f2 == f1:
            of.write(line)
        else:
            n2+=1
            print('Escrito arquivo: {}-->  {}. Possui {} registros. Linha {}'.format(n2,f1,n1-n0,n1))
            n0=n1
            of.close()
            f1 = f2
            out_file = os.path.join(OFAS,f1+'.csv')
            of = open(out_file,'wt')
print('Concluído')

    
```
