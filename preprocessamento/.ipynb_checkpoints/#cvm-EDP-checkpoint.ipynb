```python
def get_month(df, date, tolerance=10):
    begin_date = date - pd.Timedelta(days=30)
    success = False
    count = 0
    while not success and count <= tolerance:
        try:
            dfo = df[begin_date:date]
            success=True
            return dfo
        except:
            begin_date = begin_date + pd.Timedelta(days=1)
            count = count+1
    else:
        print('Mes nao preenchido')
        return None
```
