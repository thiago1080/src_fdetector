```python
# from importlib import reload
# import re
# import numpy as np
# import Fasoriais_4_2class
# reload(Fasoriais_4_2class)
# from Fasoriais_4_2class import profile, sdict
# import json
# import pandas as pd
# import os
# from genData import load_model

def get_horario_info():
    with open('../info/info.json') as jinfo:
        info = json.load(jinfo)
        HORARIOS_PATH = info['paths']['perfis']['horarios']
        PERFIS_PATH = info['paths']['perfis']['perfis']
        ATIV_PATH = info['paths']['perfis']['mapa']


    perfis = pd.read_csv(PERFIS_PATH, sep = ';')
    ativs = pd.read_csv(ATIV_PATH, sep = ';')

    dates = re.compile("[0-9]{2}:[0-9]{2}")
    dateparser = lambda x: pd.datetime.strptime(x, '%H:%M') if type(x) == str  and dates.match(x) else pd.NaT
    cols = ['name','inicio_dia_util','fim_dia_util', 'inicio_sabado', 'fim_sabado', 'inicio_domingo', 'fim_domingo', 'inicio_feriado', 'fim_feriado']
    horarios = pd.read_csv(HORARIOS_PATH, sep = ';',skipinitialspace=True, skiprows = 2, na_values = np.NaN, names = cols,
                          date_parser = dateparser, parse_dates = [1,2,3,4,5,6,7,8])
    
    
    horarios.set_index('name', inplace = True)
    horarios.dropna(inplace=True)

    perfis.set_index('CLIENTE/INSTALAÇÃO', inplace=True)
    perfis = (perfis.groupby(level=0).first())

    
    return horarios, perfis, ativs
```

```python
def kcheck(dici, key):
    if key in dici:
        return dici[key]
    else:
        return 'X'
```

```python
def keys_check(dici, keys):
    check = kcheck(dici, keys[0])
    for k in keys[1:]:
        check = kcheck(check,k)
        if check=='X':
            return 'X'
    return check
        
```

```python
def cat_out(df_out, mmo, dfo):
    for i in [1,2,3,4]:
        df_out = pd.concat( (df_out, pd.DataFrame(index = out_index, 
                                                  data = [
                                dfo['name'],
                                i,
                                mmo['empty_values'][0],
                                int(mmo['missing_rows'][0]),
                                dfo['perfil'],
                                keys_check(mmo, ['perfil',2019,i,'cat']),
                                keys_check(mmo, ['perfil',2019,i,'cats']),
                                keys_check(mmo, ['degrau',2019,i,'cat']),
                                keys_check(mmo, ['morro',2019,i,'cat']),
                                dfo[2019][i]['tension_zero']['days'],
                                dfo[2019][i]['tension_zero']['minutes'],
                                dfo[2019][i]['tension_100']['days'],
                                dfo[2019][i]['tension_100']['minutes'],
                                dfo[2019][i]['current_zero']['days'],
                                dfo[2019][i]['current_zero']['minutes'],
                                dfo[2019][i]['current_neg']['days'],
                                dfo[2019][i]['current_neg']['minutes'],
                                dfo[2019][i]['power_neg']['days'],
                                dfo[2019][i]['power_neg']['minutes'],
                                dfo[2019][i]['wrong_angles']['days'],
                                dfo[2019][i]['wrong_angles']['minutes'],  
                                dfo[2019][i]['severity']  
                            ])), axis=1, ignore_index=True )
    return df_out
```

```python
def proc_fas(fas,ddays, nn_fas, hours = ['6:00', '18:00','6:00', '18:00','0:00', '0:00','0:00', '0:00'] ):
    fas.split_months(ddays = ddays)

    for y in fas.months:
        for m in fas.months[y]:
            fas.months[y][m].remove_dup()
            fas.months[y][m].proc_fasorial()
            fas.months[y][m].feat_eng_fasorial()
            fas.months[y][m].summarize(hours = hours)                    

#     for y in fas.months:
#         for m in fas.months[y]:
#             print(y,m,'\n',fas.months[y][m].outputs)

    for y in fas.months:
        for m in fas.months[y]:
            fas.months[y][m].build_nn_input()
            fas.months[y][m].nn(nn_fas)

#     for y in fas.months:
#         for m in fas.months[y]:
#             print(fas.months[y][m].nn_input)

#     severities = {0: 'Fraude: severidade alta', 1:'Fraude: severidade moderada', 2:'Sem fraude'}

#     for y in fas.months:
#         for m in fas.months[y]:
#             print(fas.months[y][m].model_path)
#             fas.months[y][m].nn(nn_fas)
#             print('{} - {} : {}\n {}'.format(y,m, fas.months[y][m].outputs['output_nn'], fas.months[y][m].outputs['severity']))
```

```python
def processa_mm(mm, ddays,  perfil, degrau, morro):
    mm.proc_mm()
    mm.split_months(ddays = ddays, four_weeks = True)
    mm.eight_weeks(ddays = ddays)
    
    mm.nn_perfil(imodel = perfil)
    mm.degrau_full(imodel = degrau)
    mm.morro_full(imodel = morro) 
   
    
    
```
