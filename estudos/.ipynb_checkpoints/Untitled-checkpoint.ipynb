```python
import pandas as pd
%matplotlib inline

import sys
import os
from os.path import join
from pathlib import Path
sys.path.append(join(Path(os.getcwd()).parent, 'lib'))
import numpy as np
import json
from matplotlib import pyplot as plt
```

```python
with open('../../info/info-EDP-ES.json','rt') as p:
    data = json.load(p)

mm_path = data['paths']['mm']['input']
mm = pd.read_csv(mm_path, sep=';', index_col = 0, parse_dates=[0])
mm.fillna(value=0, inplace=True)
```

```python
COLS_LISTA = data['lista2']['columns']
LISTA_CAD_PATH = data['paths']['lista2']

try:
    LISTAF = pd.read_csv(LISTA_CAD_PATH, parse_dates=[5,6,7,8], names = COLS_LISTA, header=0, sep=';',engine='python')
except:
    LISTAF = pd.read_csv(LISTA_CAD_PATH, parse_dates=[5,6,7,8],names = COLS_LISTA, header=0, engine='python')
```

```python
morro = pd.read_csv('morro.csv', header=0, sep=',',engine='python')
```

```python
morro.info()
```

```python
morro = morro.drop(['Unnamed: 0'], axis=1)
```

```python
morro.sort_values(by=['instalacao'], inplace=True)
```

```python
casos_memoria_massa = pd.DataFrame(columns=['instalacao', 'data'])
for index, row in morro.iterrows():
    mes = row['mes']
    instalacao = row['instalacao']
    medidor = row['medidor']
    df_instalacao = LISTAF[(LISTAF['instalacao'] == instalacao) & (LISTAF['medidor'] == medidor)][['instalacao', mes]]
    df_instalacao.columns = ['instalacao', 'data']
    casos_memoria_massa = pd.concat([casos_memoria_massa, df_instalacao])
    
```

```python
casos_memoria_massa.drop_duplicates(inplace=True)
```

```python
casos_memoria_massa.sort_values(by=['instalacao', 'data'], inplace = True)
```
