```python
import pandas as pd
```

```python
from os import listdir as ld
```

```python
from os.path import join
```

```python
def dk(d, k=0):
    return list(d.values())[k]
```

```python
rm -rf ../outputs/all/.ipynb_checkpoints/
```

```python
d = {}
for i in ld('../outputs/all/'):
    d[i] = pd.read_csv(join('../outputs/all',i),sep=',', engine='python')
```

```python
len(d)
```

```python
for v in d.values():
    print(v.shape)
```

```python
d.keys()
```

```python
dk(d)
```

```python
df = pd.DataFrame()
for k,v in d.items():
    df = pd.concat((df,v),axis=0)
```
